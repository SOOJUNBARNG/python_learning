```python
# 1. plot() → dfを可視化
print("以下はmatplotlibで表示")
df.plot(x='date', y='price', kind='line', marker='o', title="Cost trends")
plt.xlabel("date")
plt.ylabel("price")
plt.show()
```

```python
# 2. corr() → 数値の列間の相関係数の分析
# sampledata
data = {
    'age': [25, 30, 35, 40, 45],
    'height': [160, 170, 175, 180, 185],
    'weight': [55, 65, 70, 80, 85],
    'salary': [3000, 4000, 5000, 6000, 7000]
}

df2 = pd.DataFrame(data)
print(df2)
print("\n------以下は相関係数-----\n")

# 相関係数の分析
correlation_matrix = df2.corr()
print(correlation_matrix)
```

```python
# 3. pivot() → dfの行と列を再構成する以下は給与を年連別で並べ替えたもの
pivot_df = df2.pivot_table(index='age', values='salary')
print(pivot_df)
```
